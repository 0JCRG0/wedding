```python
import pandas as pd

pd.set_option("display.max_rows", None)
pd.set_option("display.max_colwidth", None)

df = pd.read_csv("data/stats/final_guest_list.csv")

df
```

```python
import os

os.makedirs("data/stats", exist_ok=True)

print(f"Total guests: {len(df)}")
```

```python
df_entry_counts = df["entry"].value_counts().reset_index()
df_entry_counts.columns = ["entry", "count"]
df_entry_counts.to_csv("data/stats/entry_counts.csv", index=False)
df_entry_counts
```

```python
df_main_course_counts = df["main_course"].value_counts().reset_index()
df_main_course_counts.columns = ["main_course", "count"]
df_main_course_counts.to_csv("data/stats/main_course_counts.csv", index=False)
df_main_course_counts
```

```python
df_table_counts = df["table_name"].value_counts().sort_index().reset_index()
df_table_counts.columns = ["table_name", "count"]
df_table_counts.to_csv("data/stats/table_counts.csv", index=False)
df_table_counts
```

```python
df_entry_by_table = df.groupby("table_name")["entry"].value_counts().unstack(fill_value=0)
df_entry_by_table.to_csv("data/stats/entry_by_table.csv")
df_entry_by_table
```

```python
df_main_course_by_table = df.groupby("table_name")["main_course"].value_counts().unstack(fill_value=0)
df_main_course_by_table.to_csv("data/stats/main_course_by_table.csv")
df_main_course_by_table
```

```python
df_allergies = df[df["allergies"].notna()][["member", "allergies", "table_name"]]
df_allergies.to_csv("data/stats/allergies.csv", index=False)
df_allergies
```
